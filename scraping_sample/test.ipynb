```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
page_url = 'https://www.direct-store.net/blog/1722'
driver_path = r'C:\Users\User\Documents\GitHub\DS-tools\scraping_sample\chromedriver.exe'
driver = webdriver.Chrome(driver_path)
driver.get(page_url)

htmlContent = driver.page_source
soup =  BeautifulSoup(htmlContent,'html.parser')
driver.close()
print(soup)
```

```python
wanted_div = soup.findAll('div', {'class':'blog-parts'})[0]
wanted_article=wanted_div.findAll('a')
```

```python
link_list = []
for arti in wanted_article:
    link_list.append(str(arti["href"]))
    print(str(arti["href"]))
```

```python
print(link_list)
```

```python
names = "direct-store.txt"
with open(names,"w") as file:
    file_data =file.writelines(link_list[0])
print(file_data)
print(type(file_data))  

```
